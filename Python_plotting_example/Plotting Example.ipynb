## Introduction to creating beautiful plots for your LaTeX thesis:

Python is a useful tool to create publication quality plots for your thesis. This example makes use of the package [SciencePlots](https://github.com/garrettj403/SciencePlots), which creates beautiful plots. 

It also sets up the figure sizes to match the column width of the latex template. This is important to maintain consistent font sizes. Further information on why this is important can be found, and the source of some of these settings can be found  [here](https://scipy-cookbook.readthedocs.io/items/Matplotlib_LaTeX_Examples.html).

Finally, '.pdf' is the recommended filetype for saving figures for your thesis. As they're essentially vector graphics they offer a much smaller filesize while maintaining high quality when compared to image files like .png or .jpg.

```python
import numpy as np    #This imports the NumPy package we will use for this example, to plot CSV files simply google plotting from csv python to find hundreds of examples, be use to include the settings below to get the correct figure sizes
import matplotlib.pyplot as plt  #This is the package we will use to plot



plt.style.use(['science','muted','no-latex']) #This choses the SciencePlots style, see the link above for other styles and colours

fig_width_pt = 438.17247   # If changing the margins from the template settings get the new colunmwidth from LaTeX using \showthe\columnwidth
                           # The column width will be returned the log output.
    
inches_per_pt = 1.0/72.27               # Convert pt to inches
golden_mean = (np.sqrt(5)-1.0)/2.0      # Aesthetic ratio
fig_width = fig_width_pt*inches_per_pt  # width in inches
fig_height =fig_width*golden_mean       # height in inches
fig_size = [fig_width,fig_height]
SMALL_SIZE = 4
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=10)          # controls default text sizes
plt.rc('axes', titlesize=10)     # fontsize of the axes title
plt.rc('axes', labelsize=10)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=8)    # fontsize of the tick labels
plt.rc('ytick', labelsize=8)    # fontsize of the tick labels
plt.rc('legend', fontsize=8)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.figsize'] = fig_size
plt.rcParams['figure.dpi'] = 200
plt.rcParams["errorbar.capsize"] = 3
```

## Example One

Single sine wave.

Note: For a thesis plot titles are not recommended, let your caption explain the plot.

```python
x = np.linspace(0,10, num=500)  #Creates an array from 0 to 10 with 500 evenly space data points
y = np.sin(x)                   # Takes the sine of the array we just created

plt.plot(x,y,label='Sine Wave') # Plots our x and y and labels it for our legend
plt.xlabel('x')                 # Label's x
plt.ylabel('Sin(x)')            # Label's y
plt.legend()                    # Creates legend based on labels
plt.savefig('example_1.pdf')    # Saves the plot
plt.show()                      # Shows the plot in the Jupyter notebook
```

## Example Two

Multiple Lines on a plot. In this case, sine and cosine waves.

Note: For a thesis plot titles are not recommended, let your caption explain the plot.

```python
x  = np.linspace(0,10, num=500)   #Creates an array from 0 to 10 with 500 evenly space data points as before
y  = np.sin(x)                    # Takes the sine of the array we just created as before
y2 = np.cos(x)                    # Takes the cosine of the array we created

plt.plot(x,y,label='Sine Wave')   # Plots our x and y and labels it for our legend
plt.plot(x,y2,label='Cosine Wave')# Plots our second line
plt.xlabel('x')                   # Label's x
plt.ylabel('Sin(x)/Cos(x)')       # Label's y
plt.legend()                      # Creates legend based on labels
plt.savefig('example_2.pdf')      # Saves the plot
plt.show()                        # Shows the plot in the Jupyter notebook
```

### Example 5

Importing and plotting data from external files.

In this example we'll use Auger Electron Spectroscopy data which is the atomic percentage of various elements with respect to time. Given that everything is with respect to time, we'll set out index coloumn to be the time.

We'll also be using the Pandas package to import the csv file as a Pandas DataFrame. Pandas is particular useful for managing large datasets and performing operations on them. It can import multiple filetyles including Excel files and can also export it to common filetypes. See the user [guide](https://pandas.pydata.org/docs/user_guide/index.html#user-guide) for more information.


```python
import pandas as pd #first we need to import pandas

df = pd.read_csv('AES_data.csv', index_col='Sputter Time')
```

Now let's view the data we imported

```python
df
```

Here we can see there are multiple columns but the columns are named after the Auger peaks, let's rename them to the elements themselves.

```python
cols = ['', '', 'Silicon', 'Aluminium', 'Nickel', 'Oxygen']
df = df.rename(columns={"C1": "Carbon", "Ta2": "Tantalum", 'Si2':'Silicon', 'Al2':'Aluminium', 'Ni1':'Nickel','O1':'Oxygen'})
```

Now let's plot the data.

```python
df.plot() #Pandas has built in plotting
plt.show() # We still need to show the data though
```

That's not bad, but lets smooth the data a bit by instituting a moving average and improve the labels.

```python
df.ewm(span = 10).mean().plot() # we've added .ewm(span = 10).mean() to average over a range of 10 data points
plt.show() # We still need to show the data though
```

```python
df.ewm(span = 10).mean().plot(ylabel='Atomic Concentration (\%)') # Now lets add a y-label, as % has a special meaning in python and latex the \ is needed first
plt.savefig('Example5.pdf') #Let's save it as a pdf for our thesis
plt.show() # We still need to show the data though
```

We can also plot each column as a separate subplot with subplots=True, it's also useful to set the layout and choose whether axes are shared.

```python
df.ewm(span = 10).mean().plot(subplots=True,layout=(3,2), sharey=True, ylabel='Atomic Concentration (\%)') 
plt.show() # We still need to show the data though
```

As we can see the y-labels are now overlapping, and ideally we'd just have one. We've reached the limits of what's possible by df.plot and now need the full power of matplotlib, so to do that we'll initiate a subfigure with matplotlib and pass it to our dataframe, we'll also move our layout (3,2) and sharey variables to matplotlib directly. Now we can use the supylabel feature to put a single ylabel.

```python
fig, ax = plt.subplots(3,2, sharey=True)
df.ewm(span = 10).mean().plot(subplots=True, ax=ax) 
fig.supylabel('Atomic conc. (\%)')
fig.tight_layout(w_pad=2) #Adjusts the spacing between plots https://matplotlib.org/devdocs/api/figure_api.html#matplotlib.figure.Figure.tight_layout
plt.savefig('Example5_subplot.pdf') #Let's save it as a pdf for our thesis
plt.show() # We still need to show the data though
```

## Colourmap example

For data requiring a colour map, such as heatmaps, a visually uniform colour map is important. This is helpful for the colour bind and means that if your thesis is printed in black and white it'll still make sense. For this we will use a visually uniform colour map. In this case 'lajolla', though [others](https://www.fabiocrameri.ch/colourmaps/) are available. We will implement it with the python packace [cmcrameri](https://pypi.org/project/cmcrameri/)

```python
from cmcrameri import cm
fig, ax = plt.subplots()

data = np.random.rand(100,100) #create a 100 by 100 array of random data
c = ax.imshow(data,interpolation='gaussian', cmap=cm.lajolla)
fig.colorbar(c, ax=ax)
plt.savefig('Example_6_colourmaps.pdf')
plt.show()
```
